## Data cleaning

### Mixed-Type Data

```python
# finding mixed-type columns within dataframe
for col in df_orig.columns.tolist():
  weird = (df_orig[[col]].applymap(type) != df_orig[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_orig[weird]) > 0:
    print (col)
```
